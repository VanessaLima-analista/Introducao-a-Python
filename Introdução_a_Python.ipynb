```python
# Coletando dados do usuário
nome = str(input("Informe seu nome: "))
peso_str = input("Informe seu peso em kg: ")
altura_str = input("Informe sua altura em metros: ")

# Substituir vírgula por ponto nos inputs numéricos
peso = float(peso_str.replace(',', '.'))
altura = float(altura_str.replace(',', '.'))

# IMC
imc = peso / altura ** 2

# Exibindo resultado
print(f"{nome} tem IMC de {imc:.2f}")

# Classificando o IMC
if imc < 18.5:
        print("Você está abaixo do peso")
elif imc < 24.9:
        print("Você está no peso ideal")
elif imc < 29.9:
        print("Você está com sobrepeso")
else:
        print("Você está com obesidade")
```

```python
while True:
  numero = int(input("Digite um número para realizar sua tabuada: "))
  if numero < 0:
    print("Números negativos não são válidos, interrompendo o programa.")
    break
  elif numero == 0:
    print("Número 0 ignorado.")
    continue
  else:
    print(f"Tabuada do {numero}:")
    for i in range(1,11):
      resultado = numero * i
      print(f"{numero}x{i} = {resultado}")
```

```python
lista_compras = []

while True:
  print("/nOpções:)")
  print("1 - Adicionar item")
  print("2 - Remover item")
  print("3 - Mostrar a lista")
  print("4 - Sair")

  opcao = input("Escolha uma opção: ")

  if opcao == "1":
    item = input("Digite o item que deseja adicionar: ")
    lista_compras.append(item)
    print(f"{item} foi adicionado à lista de compras.")
  elif opcao == "2":
    item = input("Digite o item que deseja remover: ")
    if item in lista_compras:
      lista_compras.remove(item)
      print(f"{item} foi removido da lista de compras.")
    else:
      print("Item não encontrado na lista")
  elif opcao == "3":
    for i, item in enumerate(lista_compras, start=1):
      print(f"{i}: {item}")
  elif opcao == "4":
    print("Saindo do programa")
    break
  else:
    print("Opção não existe, escolha opção válida.")
```

```python
# Passo 1: Criar dicionário com perguntas e respostas

quiz = {
    "Qual a capital do Brasil?": "Brasília",
    "Quem desenvolveu a teoria da relatividade?": "Einstein",
    "Quantos planetas tem o sistema solar?": "8"
}

# Passo 2: Criar a lógica do quiz

acertos = 0

for pergunta, resposta in quiz.items():
  resposta_usuario = input(pergunta + " ")

  if resposta_usuario.strip().lower() == resposta.lower():
    print("Resposta correta!")
    acertos += 1
  else:
    print(f"Resposta errada! A resposta correta é: {resposta}")

print(f"Você acertou {acertos} de {len(quiz)} perguntas.")
```

```python
# Passo 1: Criando a Estrutura de Dados
usuarios = {}
proximo_id = 1  # ID inicial do usuário (variável global)

# Passo 2: Criando Funções para o Cadastro

def cadastrar_usuario(nome, idade, cidade):
    global proximo_id  # altera a variável global
    idade = int(idade)
    usuarios[proximo_id] = {"nome": nome, "idade": idade, "cidade": cidade}
    print(f"Usuário {nome} cadastrado com sucesso! ID: {proximo_id}")
    proximo_id += 1  # incrementando 1 no ID do próximo usuário

# Passo 3: Criando Funções para Analisar usuários

def listar_usuarios():
    if not usuarios:
        print("Nenhum usuário cadastrado.")
        return
    print("\nLista de usuários")
    for id, dados in usuarios.items():
        print(f"ID:{id}, Nome:{dados['nome']}, Idade:{dados['idade']}, Cidade:{dados['cidade']}")

def filtrar_por_idade(idade_minima):
    idade_minima = int(idade_minima)
    filtrados = {id: dados for id, dados in usuarios.items() if dados["idade"] >= idade_minima}
    if not filtrados:
        print("Nenhum usuário encontrado com idade maior ou igual a idade mínima.")
        return
    print(f"\nUsuários com idade maior ou igual a {idade_minima}:")
    for id, dados in filtrados.items():
        print(f"ID:{id}, Nome:{dados['nome']}, Idade:{dados['idade']}, Cidade:{dados['cidade']}")

def estatisticas_usuarios():
    if not usuarios:
        print("Nenhum usuário cadastrado.")
        return

    total = len(usuarios)
    idades = [dados["idade"] for dados in usuarios.values()]
    idade_media = sum(idades) / total
    idade_max = max(idades)
    idade_min = min(idades)

    print("\nEstatísticas dos usuários:")
    print(f"Total de usuários: {total}")
    print(f"Idade média: {idade_media:.2f}")
    print(f"Idade máxima: {idade_max}")
    print(f"Idade mínima: {idade_min}")

# Passo 4: Criando um menu interativo
def menu():
    while True:
        print("\n MENU DO SISTEMA DE USUÁRIOS.")
        print("1. Cadastrar novos usuários.")
        print("2. Listar todos os usuários.")
        print("3. Filtrar usuários por idade mínima.")
        print("4. Exibir estatísticas básicas.")
        print("5. Sair.")

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            nome = input("Nome: ")
            idade = input("Idade: ")
            cidade = input("Cidade: ")
            cadastrar_usuario(nome, idade, cidade)

        elif opcao == "2":
            listar_usuarios()

        elif opcao == "3":
            idade_minima = input("Digite a idade mínima: ")
            filtrar_por_idade(idade_minima)

        elif opcao == "4":
            estatisticas_usuarios()

        elif opcao == "5":
            print("Deslogando do sistema.")
            break

        else:
            print("Opção inválida! Tente novamente.")

menu()
```

# Simulador
A RetailCo contratou um Desenvolvedor Python para criar um sistema de gerenciamento de estoque. O objetivo é:

controlar produtos, suas quantidades e preços, e permitir a aplicação de descontos.
O sistema deve incluir funcionalidades como:

a atualização de estoque;
exibição em tempo real dos itens disponíveis;
cálculo do valor total.
O desafio é implementar um menu interativo para otimizar a gestão, com opções de:

adicionar produtos;
atualizar quantidades;
aplicar desconto;
exibir o total do estoque.

```python
# Passo 1: Criando estrutura de dados
estoque = {}

# Passo 2: Criando função para adição de produtos
def add_produto(produto, preco, quantidade):
  estoque[produto] = {"preço": preco, "quantidade": quantidade}
  print(f"\nProduto {produto} cadastrado com sucesso!")

# Passo 2: Criando função para adição de produtos
def add_quantidade(produto,quantidade_add):
  estoque[produto]["quantidade"] += quantidade_add
  print(f"\nQuantidade adicionada! Quantidade produto {produto}: {estoque[produto]['quantidade']}")

# Passo 3: Criando função para aplicar descontos
def aplicar_descontos(produto, desconto):
  preco_desconto = estoque[produto]["preço"] * (1 - desconto)
  print(f"\nDesconto aplicado! Preço do {produto} com desconto de {desconto}: {preco_desconto}")

# Passo 4: Criando função para exibir estoque
def mostra_estoque():
    print("\nEstoque:")
    for produto, dados in estoque.items():
      print(f"\nProduto: {produto}, Preço: {dados['preço']}, Quantidade: {dados['quantidade']}")

# Passo 5: Criando função para exibir valor total do estoque
def total_estoque():
  total = 0
  for produto, dados in estoque.items():
    total += dados["preço"]
  print(f"\nValor total do estoque: {total}")

# Passo 6: Menu interativo
def menu():
  while True:
    print("\n MENU DO SISTEMA DE USUÁRIOS.")
    print("1. Cadastrar novos produtos.")
    print("2. Atualizar quantidade de um produto.")
    print("3. Aplicar descontos a um produto.")
    print("4. Exibir estoque.")
    print("5. Exibir valor total do estoque.")
    print("6. Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
      produto = input("Produto: ")
      if produto in estoque:
        print("\nProduto já cadastrado! Caso deseje atualizar quantidades, escolha opção 2 do menu.")
      else:
        preco = float(input("Preço: "))
        quantidade = int(input("Quantidade: "))
        add_produto(produto, preco, quantidade)

    elif opcao == "2":
      if not estoque:
        print("\nEstoque vazio! Para adicionar um produto, selecione opção 1 do menu.")
      else:
        produto = input("Produto: ")
        if produto not in estoque:
          print("\nProduto não consta no estoque! Para adicionar um produto, selecione opção 1 do menu.")
        else:
          quantidade_add = int(input("Quantidade a adicionar ao estoque: "))
          add_quantidade(produto,quantidade_add)

    elif opcao == "3":
      if not estoque:
        print("\nEstoque vazio! Para adicionar um produto, selecione opção 1 do menu.")
      else:
        produto = input("Produto: ")
        if produto not in estoque:
          print("\nProduto não consta no estoque! Para adicionar um produto, selecione opção 1 do menu.")
        else:
          desconto = float(input("Desconto a aplicar (use formato 0.1): "))
          aplicar_descontos(produto, desconto)

    elif opcao == "4":
      if not estoque:
        print("\nEstoque vazio! Para adicionar um produto, selecione opção 1 do menu.")
      else:
        mostra_estoque()

    elif opcao == "5":
      if not estoque:
        print("\nEstoque vazio! Para adicionar um produto, selecione opção 1 do menu.")
      else:
        total_estoque()

    elif opcao == "6":
        print("Deslogando do sistema.")
        break

    else:
        print("Opção inválida! Tente novamente.")

menu()
```
